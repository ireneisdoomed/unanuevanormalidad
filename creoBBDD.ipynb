## Diccionario de las excepciones

```python
porCodigo = excep.set_index(["Provincia"])
```

```python
# Función para limpiar columna CP

def clean_CP(df):
    df["CP"] = df["CP"].apply(lambda X: X.split(","))
    df = df.explode("CP")
    df["CP"] = df["CP"].apply(lambda X: X.replace(" ", ""))
    return df
    
```

```python
# Creo diccionario de dataframes filtrados por código

df = clean_CP(porCodigo)
codigos = porCodigo.index.value_counts().index.to_list()

dfDict = {codigo : df.loc[codigo] for codigo in codigos}
```

```python
d = {}
for i in dfDict:
    
    d[i] = dfDict[i].to_dict("record")
```

```python
def parse_excepcion(nodictio,prefijo):
    dictio ={}
    dictio[prefijo] = nodictio[prefijo].to_dict("record")
    print(dictio)
    d_temp = {}
    for zona in dictio[prefijo]:
        d_temp[zona['CP']] = {'Fase':zona['Fase'],'Territorio':zona['Territorio']}

    return d_temp
```

```python
for value in data_dict:
    if value['Excepción']:
        prefijo = value["CP"]
        excepcion = parse_excepcion(dfDict,prefijo)
        value['Excepcion'] = excepcion

```
